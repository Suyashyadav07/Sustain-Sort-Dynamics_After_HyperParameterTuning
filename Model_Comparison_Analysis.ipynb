## Performance Metrics Analysis

```python
# Performance Metrics Comparison
performance_metrics = df[['Model', 'Train_Accuracy (%)', 'Test_Accuracy (%)', 'Precision', 'Recall', 'F1_Score', 'ROC_AUC']]

print("ðŸŽ¯ PERFORMANCE METRICS ANALYSIS")
print("=" * 40)
display(performance_metrics.round(3))

# Visualize performance metrics
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Model Performance Comparison', fontsize=16, fontweight='bold')

# Test Accuracy
axes[0,0].bar(df['Model'], df['Test_Accuracy (%)'], color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7'])
axes[0,0].set_title('Test Accuracy (%)')
axes[0,0].set_ylabel('Accuracy (%)')
axes[0,0].tick_params(axis='x', rotation=45)

# F1 Score
axes[0,1].bar(df['Model'], df['F1_Score'], color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7'])
axes[0,1].set_title('F1 Score')
axes[0,1].set_ylabel('F1 Score')
axes[0,1].tick_params(axis='x', rotation=45)

# ROC-AUC
axes[1,0].bar(df['Model'], df['ROC_AUC'], color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7'])
axes[1,0].set_title('ROC-AUC Score')
axes[1,0].set_ylabel('ROC-AUC')
axes[1,0].tick_params(axis='x', rotation=45)

# Training vs Test Accuracy
x = np.arange(len(df['Model']))
width = 0.35
axes[1,1].bar(x - width/2, df['Train_Accuracy (%)'], width, label='Training', alpha=0.8)
axes[1,1].bar(x + width/2, df['Test_Accuracy (%)'], width, label='Test', alpha=0.8)
axes[1,1].set_title('Training vs Test Accuracy')
axes[1,1].set_ylabel('Accuracy (%)')
axes[1,1].set_xticks(x)
axes[1,1].set_xticklabels(df['Model'], rotation=45)
axes[1,1].legend()

plt.tight_layout()
plt.show()
```

## Computational Efficiency Analysis

```python
# Computational Parameters
computational_metrics = df[['Model', 'Total_Parameters (M)', 'FLOPs (GMac)', 'Training_Time (min)', 'Model_Size_Category']]

print("âš¡ COMPUTATIONAL EFFICIENCY ANALYSIS")
print("=" * 45)
display(computational_metrics)

# Efficiency vs Performance Plot
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Computational Efficiency Analysis', fontsize=16, fontweight='bold')

# Parameters vs Accuracy
axes[0].scatter(df['Total_Parameters (M)'], df['Test_Accuracy (%)'], 
               s=100, c=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7'], alpha=0.7)
for i, model in enumerate(df['Model']):
    axes[0].annotate(model, (df['Total_Parameters (M)'][i], df['Test_Accuracy (%)'][i]), 
                    xytext=(5, 5), textcoords='offset points', fontsize=9)
axes[0].set_xlabel('Total Parameters (M)')
axes[0].set_ylabel('Test Accuracy (%)')
axes[0].set_title('Parameters vs Accuracy')
axes[0].grid(True, alpha=0.3)

# FLOPs vs Accuracy
axes[1].scatter(df['FLOPs (GMac)'], df['Test_Accuracy (%)'], 
               s=100, c=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7'], alpha=0.7)
for i, model in enumerate(df['Model']):
    axes[1].annotate(model, (df['FLOPs (GMac)'][i], df['Test_Accuracy (%)'][i]), 
                    xytext=(5, 5), textcoords='offset points', fontsize=9)
axes[1].set_xlabel('FLOPs (GMac)')
axes[1].set_ylabel('Test Accuracy (%)')
axes[1].set_title('FLOPs vs Accuracy')
axes[1].grid(True, alpha=0.3)

# Training Time
axes[2].bar(df['Model'], df['Training_Time (min)'], 
           color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7'])
axes[2].set_title('Training Time (minutes)')
axes[2].set_ylabel('Time (min)')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## Hyperparameter Tuning Results

```python
# Hyperparameter Analysis
hyperparams = df[['Model', 'Best_Learning_Rate', 'Best_Optimizer', 'Best_Batch_Size', 'Validation_Accuracy (%)']]

print("ðŸ”§ HYPERPARAMETER TUNING RESULTS")
print("=" * 40)
display(hyperparams)

# Learning Process Analysis
learning_analysis = df[['Model', 'Overfitting_Risk', 'Early_Stopping_Used', 'Transfer_Learning']]

print("\nðŸ“š LEARNING PROCESS ANALYSIS")
print("=" * 35)
display(learning_analysis)
```

## Model Robustness Analysis

```python
# Robustness Analysis
robustness_data = df[['Model', 'Test_Accuracy (%)', 'Noise_Lighting_Robustness (%)', 'Data_Augmentation']]

print("ðŸ›¡ï¸ MODEL ROBUSTNESS ANALYSIS")
print("=" * 35)
display(robustness_data)

# Calculate robustness drop
robustness_drop = df['Test_Accuracy (%)'] - df['Noise_Lighting_Robustness (%)']
robustness_comparison = pd.DataFrame({
    'Model': df['Model'],
    'Normal_Accuracy': df['Test_Accuracy (%)'],
    'Robust_Accuracy': df['Noise_Lighting_Robustness (%)'],
    'Accuracy_Drop': robustness_drop
})

print("\nðŸ“‰ ROBUSTNESS PERFORMANCE DROP")
print("=" * 35)
display(robustness_comparison)

# Visualize robustness
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Model Robustness Analysis', fontsize=16, fontweight='bold')

# Normal vs Robust Accuracy
x = np.arange(len(df['Model']))
width = 0.35
ax1.bar(x - width/2, df['Test_Accuracy (%)'], width, label='Normal Conditions', alpha=0.8)
ax1.bar(x + width/2, df['Noise_Lighting_Robustness (%)'], width, label='Noisy/Low Light', alpha=0.8)
ax1.set_title('Accuracy: Normal vs Adverse Conditions')
ax1.set_ylabel('Accuracy (%)')
ax1.set_xticks(x)
ax1.set_xticklabels(df['Model'], rotation=45)
ax1.legend()
ax1.grid(True, alpha=0.3)

# Accuracy Drop
ax2.bar(df['Model'], robustness_drop, color=['#E74C3C', '#3498DB', '#2ECC71', '#F39C12', '#9B59B6'])
ax2.set_title('Accuracy Drop in Adverse Conditions')
ax2.set_ylabel('Accuracy Drop (%)')
ax2.tick_params(axis='x', rotation=45)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Error Analysis - Confusion Matrix Insights

```python
# Confusion Matrix Analysis
confusion_data = df[['Model', 'CM_TN_NonDeg', 'CM_FP_NonDeg', 'CM_FN_Deg', 'CM_TP_Deg']]

print("ðŸ” CONFUSION MATRIX ANALYSIS")
print("=" * 35)
print("TN: True Negative (Non-Degradable correctly classified)")
print("FP: False Positive (Degradable misclassified as Non-Degradable)")
print("FN: False Negative (Non-Degradable misclassified as Degradable)")
print("TP: True Positive (Degradable correctly classified)")
print()
display(confusion_data)

# Calculate error rates
total_samples = df['CM_TN_NonDeg'] + df['CM_FP_NonDeg'] + df['CM_FN_Deg'] + df['CM_TP_Deg']
false_positive_rate = (df['CM_FP_NonDeg'] / (df['CM_FP_NonDeg'] + df['CM_TN_NonDeg'])) * 100
false_negative_rate = (df['CM_FN_Deg'] / (df['CM_FN_Deg'] + df['CM_TP_Deg'])) * 100

error_analysis = pd.DataFrame({
    'Model': df['Model'],
    'Total_Samples': total_samples,
    'False_Positive_Rate (%)': false_positive_rate.round(2),
    'False_Negative_Rate (%)': false_negative_rate.round(2),
    'Total_Errors': df['CM_FP_NonDeg'] + df['CM_FN_Deg']
})

print("\nâŒ ERROR RATE ANALYSIS")
print("=" * 25)
display(error_analysis)
```
